```python
import math

def calculate_sunrise_sunset(latitude, longitude, year, month, day):
    # Convert longitude to degrees from 0 to 360
    if longitude < 0:
        longitude += 360
    
    # Convert latitude to radians
    latitude_rad = math.radians(latitude)
    
    # Determine the day of the year
    n1 = math.floor(275 * month / 9)
    n2 = math.floor((month + 9) / 12)
    n3 = (1 + math.floor((year - 4 * math.floor(year / 4) + 2) / 3))
    day_of_year = n1 - (n2 * n3) + day - 30
    
    # Calculate the Sun's mean anomaly
    mean_anomaly = 0.9856 * day_of_year - 3.289
    
    # Calculate the Sun's true longitude
    sun_longitude = mean_anomaly + (1.915 * math.sin(math.radians(mean_anomaly))) + (0.020 * math.sin(math.radians(2 * mean_anomaly))) + 282.634
    
    # Adjust the Sun's true longitude
    if sun_longitude > 360:
        sun_longitude -= 360
    
    # Calculate the Sun's right ascension
    right_ascension = math.degrees(math.atan(0.91764 * math.tan(math.radians(sun_longitude))))
    
    # Adjust the Sun's right ascension
    right_ascension = right_ascension % 360
    
    # Calculate the Sun's declination
    declination = math.asin(0.39782 * math.sin(math.radians(sun_longitude)))
    
    # Calculate the Sun's local hour angle
   # Calculate the Sun's local hour angle
    cos_hour_angle = (math.sin(math.radians(-0.8333)) - math.sin(latitude_rad) * math.sin(declination)) / (math.cos(latitude_rad) * math.cos(declination))
    
    if cos_hour_angle > 1 or cos_hour_angle < -1:
        # The Sun never rises or sets on the given day
        return None, None
    
    # Calculate the Sun's local mean time of rising and setting
    hour_angle = math.degrees(math.acos(cos_hour_angle))
    sunrise_time = 12 - (hour_angle / 15)
    sunset_time = 12 + (hour_angle / 15)
    
    # Format the time strings
    sunrise_str = format_time(sunrise_time)
    sunset_str = format_time(sunset_time)
    
    return sunrise_str, sunset_str

def format_time(time):
    hour = int(time)
    minute = int((time - hour) * 60)
    return f"{hour:02d}:{minute:02d}"

# Example usage:
latitude = 27.7749  # Latitude of San Francisco
longitude = 85.4194  # Longitude of San Francisco
# year = 2023
# month = 7
# day = 13


year = int(input("Enter the year:"))
month = int(input("Enter the month:"))
day = int(input("Enter the day"))



sunrise, sunset = calculate_sunrise_sunset(latitude, longitude, year, month, day)
if sunrise is not None and sunset is not None:
    print(f"Sunrise: {sunrise}")
    print(f"Sunset: {sunset}")
else:
    print("The Sun never rises or sets on the given day.")
```
